```python
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
regression.get_params()
```

```python
reg_prid= regression.predict(X_test)
```

```python
reg_prid
```

#Assumptions

```python
plt.scatter(y_test,reg_prid)
```

```python
residuals=y_test-reg_prid
```

```python
residuals
```

```python
sns.displot(residuals,kind="kde")
```

```python
plt.scatter(reg_prid,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_prid))
print(mean_squared_error(y_test,reg_prid))
print(np.sqrt(mean_squared_error(y_test,reg_prid)))
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_prid)
print(score)
```
